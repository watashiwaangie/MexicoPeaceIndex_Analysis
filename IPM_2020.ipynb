## Implementación de K-Means

#### Visualizando el DataFrame una vez más

```python
df.head(5)
```

#### Creando otro df donde quitemos los datos de tipo object

```python
DF = df.drop(['Estados', 'ID'], axis = 1)
DF.describe().T
```

#### Normalizando los datos

```python
DF_norm = (DF-DF.min())/(DF.max()-DF.min())
DF_norm.describe().T
```

#### Buscando la cantidad óptima de clusteres

Método del codo

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    # Aplicando k-means al df
    kmeans.fit(DF_norm) 
    wcss.append(kmeans.inertia_)
wcss
```

Graficando los resultados anteriores para aplicar el método del codo 

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,11), wcss, 'go--')
plt.title('El codo de Jambú')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show
```

Vemos que el número de clusteres óptimo es n = 3, ya que la atenuación es menor a partir de este punto.

#### Aplicando el método anterior al DF, con n = 3

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(DF_norm)
y_kmeans
```

```python
# Cambiamos los valores de los grupos para comenzar en 1 y lo agregamos al dataframe original
df['cluster'] = y_kmeans + 1
df.head()
```

```python
# Comparando a pares las variables
fig = plt.figure(figsize=(10, 5))
sns.pairplot(df, hue="cluster",  palette=['blue','yellow','red'], height=1.7, aspect=1)
plt.show()
```

Número de clústeres aplicando el método de la silueta.

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
#
silhouette_score(DF_norm, kmeans.labels_)
```

```python
# Cada elemento de la lista generada es un modelo Kmeans con un número específico de clústeres.
kmeans_sil= [KMeans(n_clusters= i, random_state= 42, init='k-means++').fit(DF_norm) for i in range(1,11)]

# Cálculo del silhouette score a cada modelo (de diferente número de clústeres).
sil_scores = [silhouette_score(DF_norm, model.labels_, random_state=11) for model in kmeans_sil[1:]]

# Guardando los valores en un array:
sil_scores = np.array(sil_scores)
sil_scores
```

```python
sns.set( style = 'white' )
plt.rc('font', family='serif')

plt.figure( figsize=(10.5,6),
           dpi = 150,
           facecolor=c_background
          )
plt.axes().set_facecolor = c_background

plt.plot(range(2, 11), sil_scores,
         linestyle = ':', linewidth = 1.5,
         marker = 'o', markersize = 10,
         color = '#2F968B' ,
         label = 'The Silhoutte Method'
        )


plt.title('The Silhouette Method', size = 25, pad = 10)
plt.xlabel('Number of clusters', size = 20)
plt.ylabel('Silhoutte Coefficent', size = 20)
plt.tick_params( labelsize = 15 )

plt.grid( b = True, linestyle = 'dashed', alpha = 0.5 )

plt.show()

```

```python
print( 'Max. Silhouette Score:', sil_scores[ np.argmax(sil_scores) ] )
print( 'Optimal number of cluters:', np.argmax(sil_scores) + 1 + 1 )
```
