# Observar correlación entre variables

```python
df_correlations = df.corr()
df_correlations.style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1)
```

```python
mask = np.zeros_like(df_correlations.values)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize = (8,6),
                       #dpi = 150
                      )

sns.heatmap(data=df_correlations, cmap='coolwarm', square= False, #mask=mask,
            vmin = -1, vmax = 1, center = 0,
            cbar = True, 
            annot = True, fmt = '.2f',
            annot_kws = {'size':13, 'fontweight':'bold'},
           )
ax.tick_params(axis='both', length = 0)
ax.tick_params(axis='both', rotation = 0, pad = 20, labelsize = 17)
ax.tick_params(axis='x', rotation=90)
fig.suptitle('Correlation matrix', y = 1.0, size = 23)
#fig.text(x=0.5, y=0.91, s="Lower correlated variables", ha = 'center',
#         fontdict = {'fontstyle':'italic', 'size': 16})
```

***

```python
corr_matrix = df.corr()
df_dummy = pd.get_dummies(df)
df_aux = pd.concat([df, df_dummy], axis = 1)
```

```python
mask_aux = np.zeros_like(corr_matrix, dtype=np.bool)
mask_aux[np.triu_indices_from(mask_aux)]= True
```

```python
f, ax = plt.subplots(figsize=(11, 15)) 
#sns.set()

heatmap = sns.heatmap(corr_matrix, 
                      mask = mask_aux,
                      #font_scale=1.8,
                      square = True,
                      linewidths = .5,
                      cmap = 'coolwarm',
                      cbar_kws = {'shrink': .6, 
                                'ticks' : [-1,- 0.5, 0,0.5, 1]},
                      vmin = -1, 
                      vmax = 1,
                      annot = True,
                      annot_kws = {'size': 15})#add the column names as labels
ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
#plt.ylabel('EBW (kg)', fontsize=13)
#fig = sns.get_figure()
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

**COMENTARIOS (Isaac)**

Una manera de _mejorar_ lo que hicimos la última vez, es _eliminar_ variables fuertemente correlacionadas ya que es puede llegar a ser redundante. Las formas y _patrones_ se pueden observar cuando la correlación es muy baja.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Correlation_examples2.svg/800px-Correlation_examples2.svg.png" width=700px></img>


Algunas maneras de hacer esta tarea es, literlamente dejarlas fuera de la implementación del algoritmo o juntarlas en una sola _nueva_ variable, por ejemplo, usar **PCA** para crear una _variable artificial_ (pasar de **N** variables a solamente una) que resuma la información de, por ejemplo:
* **Delitos_con_armas_fuego**
* **Homicidio**
* **Impacto_economico_per_capita_de_la_violencia**

Que son variables que están fuertemente correlacionadas entre si, con correlaciones positivas (también puede aplicar a variables con correlaciones negativas igual de fuertes)

Lo que se tiene que discutir y probar, será ese _threshold_ de correlación
